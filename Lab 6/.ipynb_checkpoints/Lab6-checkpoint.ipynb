# Реалізація алгоритму Евкліда для знаходження найбільшого спільного дільника за допомогою ділення та віднімання. 

```python
def gcd_combined(a, b):
    while a != b:
        if a > b:
            if a % b == 0:
                return b
            else:
                a -= b
        else:
            if b % a == 0:
                return a
            else:
                b -= a
    return a
```

```python
def task_2(*args):
    if len(args) != 2:
        print("Помилка. Потрібно ввести рівно 2 числа.")
        return
    
    try:
        a, b = int(args[0]), int(args[1])

        gcd_result = gcd_combined(a, b)
        print(f"Найбільший спільний дільник {a} і {b} - це {gcd_result}")
    except ValueError:
        print("Помилка. У списку присутнє нечислове значення.")
```

```python
task_2(16, 36)
```

```python
task_2(12, 54)
```

```python
task_2(3, "str")
```

```python
task_2(12, 34, 56)
```

```python
task_2(12)
```
